```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
from scipy.io import wavfile


fs, data = wavfile.read('/content/voice.wav')

def plot_fft(signal, s, title):
    n = len(signal)
    freq = np.fft.fftfreq(n, d=1/s)
    magnitude = np.abs(np.fft.fft(signal))
    plt.figure(figsize=(8, 4))
    plt.plot(freq[:n//2], magnitude[:n//2])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title(title)
    plt.grid(True)
    plt.show()


# FFT for original signal
plot_fft(data, s, 'Original Signal FFT')
```

```python
#vs is vowel is samples
# A1. FFT for vowel sounds
vs = [slice(start, start + s) for start in range(0, len(data), s)]
for idx, sample in enumerate(vs):
    plot_fft(data[sample], s, f'Vowel Sound {idx + 1}')
```

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
from scipy.io import wavfile
import IPython.display as ip
import librosa
import librosa.display


s, data = wavfile.read('/content/voice.wav')

def plot_fft(signal, s, title):
    n = len(signal)
    freq = np.fft.fftfreq(n, d=1/s)
    magnitude = np.abs(np.fft.fft(signal))
    plt.figure(figsize=(8, 4))
    plt.plot(freq[:n//2], magnitude[:n//2])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title(title)
    plt.grid(True)
    plt.show()


plot_fft(data, s, 'Original Signal FFT')


def play_audio(segment, rate):
    signal_ifft = np.fft.ifft(segment)
    ip.display(ip.Audio(signal_ifft.real, rate=rate))







```

```python
vs = [slice(start, start + s) for start in range(0, len(data), s)]
for idx, sample in enumerate(vs):
    fft_result = np.fft.fft(data[sample])
    play_audio(fft_result, s)
    magnitude_spectrum = np.abs(fft_result)
    n = len(data[sample])
    freq = np.fft.fftfreq(n, d=1/fs)  # Assuming fs is the sampling frequency

# Plot the magnitude spectrum
    plt.figure(figsize=(8, 4))
    plt.plot(freq[:n//2], magnitude_spectrum[:n//2])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.title('FFT Result od Vowels')
    plt.grid(True)
    plt.show()
```

```python
cs = [slice(start, start + s) for start in range(s//2, len(data), s)]
for idx, sample in enumerate(cs):
    fft_result = np.fft.fft(data[sample])
    play_audio(fft_result, s)
    magnitude_spectrum = np.abs(fft_result)
    n = len(data[sample])
    freq = np.fft.fftfreq(n, d=1/fs)  # Assuming fs is the sampling frequency

# Plot the magnitude spectrum
    plt.figure(figsize=(8, 4))
    plt.plot(freq[:n//2], magnitude_spectrum[:n//2])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.title('FFT Result of consonants')
    plt.grid(True)
    plt.show()
```
