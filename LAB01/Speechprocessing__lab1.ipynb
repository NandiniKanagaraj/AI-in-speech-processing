```python
import IPython.display as ipd
import pandas as pd
from glob import glob
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
af = glob("/content/WhatsApp Audio 2024-02-05 at 13.37.51.wav")
ipd.Audio(af[0])
```

```python
#file is the input audio file
file = '/content/WhatsApp Audio 2024-02-05 at 13.37.51.wav'

y, sr = librosa.load(file)
#displaying the waveform of audio
librosa.display.waveshow(y)
```

## A2. Observe the length & magnitude range of the signalfrom theplot. Observing the plot, relate tothe spoken words and silence between the words.

```python
print(f'y: {y[: 10]}')
print(f'shape y: {y.shape}')
print(f'sr:{sr}')
```

## A3. Take a small segment of the signal and play it.

```python
#segment1
st = 0.5
et = 1
si = int(st*sr)
ei = int(et*sr)

segmentation = y[si:ei]
librosa.display.waveshow(segmentation, sr=sr)

```

```python
#segment2
st1 = 1
et1 = 1.5
si1 = int(st1*sr)
ei1 = int(et1*sr)

segmentation1 = y[si1:ei1]
librosa.display.waveshow(segmentation1, sr=sr)
```

## A4. Play around with your recorded speech signal for various segments. Understand the nature of the signal. Also observe with abruptly segmented speech, how the perception of that speech is affected.

```python
# Trimming leading/lagging silence
# here the input signal is trimmed to 20 decibels
y_trimmed, _ = librosa.effects.trim(y, top_db=20)
pd.Series(y_trimmed).plot(figsize=(5,10 ),
                  lw=1,
                  title='Raw Audio Trimmed Exampl')
plt.show()
```

```python
# here the input audio file is zoom for clear visualisation

pd.Series(y[30000:30500]).plot(figsize=(10, 5),
                  lw=1,
                  title='Raw Audio Zoomed In Example')
plt.show()
```
